# Using Numpy Analyzing Weather Data

## Load weather data and compute basic information
weather data url:
http://www.sciamachy-validation.org/climatology/daily_data/selection.cgi

```python
import numpy as np
import numpy.ma as ma
from datetime import datetime as dt

to_float = lambda x: float(x.strip() or np.nan)
def to_year(x):
    x = str(x)
    x = x[2:len(x)-5]
    x = float(x)
    return x


#Measurements are in tenths of degrees
years,avg_temp,min_temp, max_temp = np.loadtxt("KNMI_20180101.txt.tsv", delimiter=',', 
                                   usecols=(1, 11, 12, 14), unpack=True,
                                   converters={1: to_year, 11:to_float, 12: to_float, 14: to_float})

avg_temp = .1 * avg_temp
min_temp = .1 * min_temp
max_temp = .1 * max_temp
print("****************************************************")
print("Freeze range:")
print('%.1f percent of days min tempurature below zero '%(len(min_temp[min_temp < 0])/len(min_temp)*100))
print('%.1f percent of days max tempurature below zero '%(len(max_temp[max_temp < 0])/len(max_temp)*100))

ranges = max_temp - min_temp
print("****************************************************")
print("Daily range:")
print("maximum daily range: ", np.nanmax(ranges))
print("minimum daily range: ", np.nanmin(ranges))

# mask array using for processing missing value
masked_ranges = ma.array(ranges,mask = np.isnan(ranges))
masked_mins = ma.array(min_temp, mask = np.isnan(min_temp))
masked_maxs = ma.array(max_temp, mask = np.isnan(max_temp))
masked_avgs = ma.array(avg_temp, mask = np.isnan(avg_temp))
print("****************************************************")
print("Average infomation:")
print("Average daily range: ", masked_ranges.mean())
print("Standard deviation: ", masked_ranges.std())
print("Average maximum temp:", masked_maxs.mean())
print("Standard deviation: ", masked_maxs.std())
print("Average maximum temp:", masked_mins.mean())
print("Standard deviation: ", masked_mins.std())
```

## Showing evidence of Global Warming

```python
%matplotlib inline
import matplotlib.pyplot as plt

print("First year: %s, Last year: %s"%(years[0],years[-1]))

year_temp_avg = []
year_range = range(int(years[0]),int(years[-1])+1)

for year in year_range:
        indices = np.where(years == year)
        year_temp_avg.append(avg_temp[indices].mean())

plt.figure(figsize=(10,5))
plt.plot(year_range, year_temp_avg, 'g-', label="Yearly Averages")
plt.plot(year_range, np.ones(len(year_temp_avg)) * np.mean(year_temp_avg))
plt.legend(prop={'size':'large'})
plt.show()
```
